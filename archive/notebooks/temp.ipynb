```python
def deduplicate_bib(bib_content):
    # Split on '@article{'
    chunks = bib_content.split('@article{')

    seen = set()
    results = []

    for chunk in chunks:
        chunk = chunk.strip()
        # Skip any empty splits
        if not chunk:
            continue

        # The piece starts like "Key, ... }"
        # Find the key before the first comma
        idx = chunk.find(',')
        if idx == -1:
            # Malformed or no comma => skip
            continue

        key = chunk[:idx].strip()
        # Reconstruct the entry
        entry = '@article{' + chunk

        # Keep the first occurrence of each key
        if key not in seen:
            seen.add(key)
            results.append(entry)

    return "\n\n".join(results)

# Example usage:
with open("input.bib", "r") as f:
    bib_text = f.read()

deduped = deduplicate_bib(bib_text)

with open("output.bib", "w") as f:
    f.write(deduped)
```
